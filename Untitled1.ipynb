```python
# Convert categorical columns to numerical using one-hot encoding
df = pd.get_dummies(df, columns=['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], drop_first=True)

# Now calculate the correlation
correlation_matrix = df.corr()

# Display the correlation matrix
print(correlation_matrix)
```

```python
#classifying patients according to chest pains
chest_pain_mapping = {
    'ATA': 'Atypical Angina',
    'NAP': 'Non-Anginal Pain',
    'TA': 'Typical Angina',
    'ASY': 'Asymptomatic'
}
df['ChestPainCategory'] = df['ChestPainType'].map(chest_pain_mapping)
print(df[['ChestPainType', 'ChestPainCategory']].head())
chest_pain_counts = df['ChestPainCategory'].value_counts()
print("\nChest Pain Category Counts:")
print(chest_pain_counts)
```

496 patients are classified as 'Asymptomatic' (meaning they don't experience typical chest pain symptoms).
203 patients have 'Non-Anginal Pain' (chest pain not related to heart disease).
173 patients have 'Atypical Angina' (chest pain that doesn't fit the classic pattern of angina).
46 patients have 'Typical Angina' (chest pain that is a classic symptom of heart disease).


```python
plt.figure(figsize=(20,10))
plt.title("chest pain of patients")
sns.barplot(x=chest_pain_counts.index, y=chest_pain_counts.values)
```

```python
import kagglehub
import pandas as pd


path = kagglehub.dataset_download("fedesoriano/heart-failure-prediction")

print("Path to dataset files:", path)

# Loading  the dataset into a pandas DataFrame called 'df'
df = pd.read_csv(path + '/heart.csv')

# Now you can display the first few rows using df.head()
df.head()
```

lets split the data into Dependent ie our target variable and independent variable


```python
x=df.iloc[:,:-1]
x
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder



y = df.iloc[:, -1]
lbl = LabelEncoder()
y = lbl.fit_transform(y)
y
```

import Eval ML Library

```python
import evalml
```

standardisation


```python
x_train,x_test ,y_train , y_test = evalml.preprocessing.split_data(x,y,problem_type='binary')
```

```python
evalml.problem_types.ProblemTypes.all_problem_types
```

running Auto ML to select the best Algorith

```python
from evalml.automl import AutoMLSearch
automl = AutoMLSearch(X_train=x_train, y_train=y_train, problem_type='binary')
automl.search()
```

```python
automl.rankings
```

```python
automl.best_pipeline
```

```python
automl.describe_pipeline(automl.rankings.iloc[0]["id"])
```

```python

best_pipeline = automl.best_pipeline


best_pipeline.score(x_test, y_test, objectives=["auc", "f1", "Precision", "Recall"])
```

Now if we want to build our model for a specific objective we can do that

```python
automl_auc = AutoMLSearch(X_train=x_train, y_train=y_train,
                          problem_type='binary',
                          objective='auc',
                          additional_objectives=['f1', 'precision'],
                          max_batches=1,
                          optimize_thresholds=True)

automl_auc.search()
```

```python
automl_auc.rankings
```

```python
automl_auc.describe_pipeline(automl_auc.rankings.iloc[0]["id"])
```

```python
best_pipeline_auc = automl_auc.best_pipeline
```

```python
#get the score on on holdout data
best_pipeline_auc.score(x_test, y_test,  objectives=["auc"])
```

we got 94.5% which is the highest of all

save the model

```python
best_pipeline.save("model.pkl")
```

loading our model

```python
final_model = automl.load('model.pkl')
```

```python
final_model.predict_proba(x_test)
```
